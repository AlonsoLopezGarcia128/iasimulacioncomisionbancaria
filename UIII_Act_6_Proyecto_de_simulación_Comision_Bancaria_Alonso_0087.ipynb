```python
print("=== SISTEMA INTELIGENTE DE COMISIONES BANCARIAS ===")
print("Bienvenido al asistente virtual del banco\n")

salir = "no"

while salir != "si":

    print("\nSeleccione el tipo de operación:")
    print("1. Retiro")
    print("2. Transferencia")
    print("3. Consulta de saldo")

    opcion = input("Ingrese una opción (1-3): ")

    if opcion == "1":
        monto = float(input("Ingrese el monto a retirar: "))

        if monto < 1000:
            comision = monto * 0.01
        elif monto >= 1000 and monto <= 5000:
            comision = monto * 0.02
        else:
            comision = monto * 0.03

        total = monto + comision

        print("\n--- RESULTADO ---")
        print("Tipo: Retiro")
        print("Monto:", monto)
        print("Comisión:", comision)
        print("Total a descontar:", total)

    elif opcion == "2":
        monto = float(input("Ingrese el monto a transferir: "))

        if monto < 2000:
            comision = 10
        elif monto >= 2000 and monto <= 7000:
            comision = 20
        else:
            comision = 50

        total = monto + comision

        print("\n--- RESULTADO ---")
        print("Tipo: Transferencia")
        print("Monto:", monto)
        print("Comisión fija:", comision)
        print("Total:", total)

    elif opcion == "3":
        print("\n--- RESULTADO ---")
        print("Consulta de saldo")
        print("Comisión: $0 (Gratis)")

    else:
        print("\nOpción inválida. Intente de nuevo.")

    salir = input("\n¿Desea salir? (si/no): ")

print("\nGracias por usar el sistema inteligente del banco.")
```
